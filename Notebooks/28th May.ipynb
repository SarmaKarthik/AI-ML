### Problem 3: 
#### Problem Statement :
 Print number of matches won and loss(as two columns). Consider as win when a team wins Toss  and Match . print for each and every team


```python
import pandas as pd
import numpy as np


df = pd.read_csv('DataFiles/Match.csv')

teams = np.unique(df.Team1)

df1 = pd.DataFrame()

df1['Teams'] = teams

w = []
l = []

for x in teams:
    
    win = loss = 0
    for i in range (0,len(df)):
        if x == df.Toss_Winner[i] and x == df.match_winner[i] :
            win = win + 1
        elif x == df.Team1[i] or x == df.Team2[i] :
            loss = loss + 1
    w.append(win)
    l.append(loss)
    
df1['win'] = w
df1['loss'] = l

df1
```
