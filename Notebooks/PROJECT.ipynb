##                               IMPORTING THE LIBRARIES                             

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## READING THE DATA-SET

```python
df=pd.read_csv('train.csv')
df.head()
```

# HANDLING MISSING VALUES

```python
df=df.dropna(how='all')
```

```python
df=df.fillna(0)
df.tail()
```

# CATEGORICAL DATA

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
dfle=df
dfle.Gender=le.fit_transform(df.Gender)
dfle.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
dfle=df
dfle.Age=le.fit_transform(df.Age)
dfle.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
dfle=df
dfle.Product_ID=le.fit_transform(df.Product_ID)
dfle.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
dfle=df
dfle.City_Category=le.fit_transform(df.City_Category)
dfle.head()
```

```python
df=df.drop(['User_ID','Product_ID','Stay_In_Current_City_Years'],axis=1)
df
```

# BUILDING THE MODEL

```python
import statsmodels.formula.api as smf
df1=smf.ols('Purchase~Gender+Age+Occupation+City_Category+Marital_Status+Product_Category_1+Product_Category_2+Product_Category_3',data=df).fit()
```

```python
df1.summary()
```

```python
import statsmodels.api as sm
sm.graphics.plot_partregress_grid(df1)
```

# SPLITTING AND TESTING

```python
X=df.iloc[:,0:8]
X.head()
```

```python
y=df.iloc[:,8]
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(X_train,y_train)
```

```python
pred=reg.predict(X_test)
pred
```

```python
pred==y_test
```
